```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

a1 = -3.983035
a2 = 301.797
a3 = 522528.9
a4 = 69.34881
a5 = 999.974950
T = np.arange(0,15,0.5)

#print T

rho=a5*(1-(T+a1)**2*(T+a2)/(a3*(T+a4)))



plt.plot(T,rho)
plt.ylabel('Density [kg/m^3]');
plt.xlabel('Temperature [$^\circ$C]');
title('Density of water per temperature')
#plt.show()

```
